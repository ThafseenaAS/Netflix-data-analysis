```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Netflix Data Analysis

## Objective

The objective of this project is to analyze Netflix Movies and TV Shows data to identify content trends, country contributions, genres, datings, and content growth over time.

## Data Loading

```python
df=pd.read_csv("netflix_titles.csv")
df.head()
```

## Data Understanding

```python
df.shape
```

```python
df.columns
```

```python
df.info()
```

```python
df.isnull().sum()
```

## Data Cleaning

```python
df['country']=df['country'].fillna("Unknown")
df['director']=df['director'].fillna("Not Available")
df['cast']=df['cast'].fillna("Not Available")
df['rating']=df['rating'].fillna("Not Rated")
df['date_added']=pd.to_datetime(df['date_added'],errors='coerce')
df['date_added']=df['date_added'].ffill()

```

```python
df.isnull().sum()
```

## Movies vs TV Shows Analysis

```python
df['type'].value_counts()
```

```python
df['type'].value_counts().plot(kind='bar')
plt.title("Movies vs TV Shows")
plt.show()
```

## Insight 1

Netflix contains significantly more Movies (6131) than TV Shows (2676) 

## Top 10 Countries on Netflix

```python
df['country'].value_counts().head(10)
```

```python
df['country'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Countries on Netflix")
plt.show()
```

## Insight 2

Most Netflix content comes from the United States, followed by India and the United Kingdom.

## Netflix Content Growth Analysis

```python
df['release_year'].value_counts().sort_index().plot(kind='line')
plt.title("Netflix Content by Release Year")
plt.show()
```

## Insight 3

Netflix content releases increased significantly over the years and peaked around 2019-2020.

## Genre Analysis

```python
df['listed_in'].value_counts().head(10).plot(kind='bar')
plt.title("Top Netflix Genres")
plt.show()
```

## Insight 4

Most Netflix content belongs to genres like Dramas, International Movies, and Comedies, indicating strong focus on entertainment and global content.

## Top 10 Release Years Analysis

```python
df['release_year'].value_counts().head(10)
```

```python
df['release_year'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Release Years")
plt.show()
```

## Insight 5

2018 had the highest number of titles (1147), followed by 2017 and 2019.

## Content Rating Analysis

```python
df['rating'].value_counts()
```

```python
df['rating'].value_counts().head(10).plot(kind='bar')
plt.title("Content Rating Distribution")
plt.show()
```

## Insight 6

TV-MA and TV-14 are among the most commom ratings on Netflix.

## Conclusion

This project analyzed Netflix content using Python, Pandas, and Matplotlib. Data cleaning was performed to handle missing values. Exploratory analysis revealed that Movies dominate the platform, the United States contributes the most content, and Netflix content growth increased significantly over the years.
